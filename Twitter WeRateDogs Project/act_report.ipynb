###                         WeRateDogs Twitter Project

The project was meant to rate peoples dogs for likes from images and videos posted on twitter. After wrangling the twitter data, making it ready for analysis, The consolidated tables that had been saved as twitter_archive_master was read into the jupyter notebook. The average rating ratio was estimated to be 12/10 using the mean and round function being applied on rating_numerator column.
It would also be great to see a trend for twitter likes by Month. The result of this is a sum all the Twitter Likes in a month and see how they faired on a monthly basis. The groupby function was used in this instance with the Month column as the argument while the sum function was applied on the Twitter Likes. The Month column was created from the timestamp column to support this analysis. The outcome of the filteration was plotted with the month number on the x-axis and Likes on the y-axis.
The dataframe was also filtered to find the Dog with the highest likes with a Yorkshire Terrier named Leo scoring the highest likes with a rating of 13/10.



A scatter plot was created to find the correlation between twitter Likes as an dependent variable and rating_numerator, p1_conf, image_number and retweets as independent variables. The initial outcome showed presence of outliers and after they were corrected for and replotted. The later outcome showed no correlation between the identified variables. The chart showed the data is concentrated between rating numerators of 0 and 14.
